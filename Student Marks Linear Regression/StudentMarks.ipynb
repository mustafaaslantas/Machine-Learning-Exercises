```python
#Gerekli kütüphaneler
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
#Veri setimizi çağıralım
df = pd.read_csv("Student_Marks.csv")
```

```python
#Veri seti bilgisi
print(df.info())
```

```python
#Verileri gözden geçirelim
df.head(5)
```

```python
#Bağımsız değişkenleri tanımlayalım
y = df[["Marks"]]
x = df[["number_courses", "time_study"]]
```

```python
#Lineer Regression Modelini Oluşturalım
l = LinearRegression()
#Modeli eğitelim
model = l.fit(x.values,y.values)
```

```python
#Katsayılar
model.coef_
```

```python
#Sabit
model.intercept_
```

```python
'''Tahminde bulunalım 
(Günde 4 saat ders çalışan 4. sınıf öğrencisi tahmini ne kadar not alır)'''
model.predict([[4,5]])
```

```python
#Sağlamasını yapalım
(4*1.86405074)+(5*5.39917879)+(-7.45634623)
```

```python
#Hata payı ve doğruluk skoru
df.head(1)
```

```python
#Tablodaki ilk öğrenciyi hesaplayalım
s1 = model.predict([[3,4.508]])[0][0] 
print(s1)
```

```python
# 3. sınıfa giden ve günde 4.508 saat çalışan bir öğrencinin 22.475 puan alması gerekirken 19.202 puan almış. 
# Buna göre hata payı hesabımızı yapalım.
```

```python
#Hata payı hesabı
(19.202-s1)/19.202
```

```python
#Modelin doğruluk skoru
model.score(x.values,y.values)
```
