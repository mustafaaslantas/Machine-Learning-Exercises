```python
#Gerekli kütüphaneler
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
#Veri setimizi çağıralım
df = pd.read_csv("Student_Marks.csv")
```

```python
#Veri seti bilgisi
print(df.info())
```

```python
#Verileri gözden geçirelim
df.head(5)
```

```python
#Bağımsız değişkenleri tanımlayalım
y = df[["Marks"]]
x = df[["number_courses", "time_study"]]
```

```python
#Lineer Regression Modelini Oluşturalım
l = LinearRegression()
#Modeli eğitelim
model = l.fit(x.values,y.values)
```

```python
'''Tahminde bulunalım 
(Günde 4 saat ders çalışan 4. sınıf öğrencisi tahmini ne kadar not alır)'''
model.predict([[4,4]])
```
